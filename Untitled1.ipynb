```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

spark = (
            SparkSession
                .builder
                .appName("AqeDynamicJoinApp")
                .master("local[4]")    
                .config("spark.dynamicAllocation.enabled","false")
                .config("spark.sql.adaptive.enabled","false")
                .getOrCreate()
        )

sc = spark.sparkContext

spark
```

```python
YellowTaxiDF = (spark.read.option("header","true").option("inferSchema","true").csv("C:\DataFiles\YellowTaxis_202210.csv"))

YellowTaxiDF.printSchema()
# YellowTaxiDF.count()
```

```python
YellowTaxiDF.createOrReplaceTempView("YellowTaxis1")
YellowTaxiDF.createOrReplaceTempView("YellowTaxis2")
```

```python
spark.sql(""" SELECT * FROM YellowTaxis1 yt1 
                JOIN 
                    YellowTaxis2 yt2
                    ON yt1.PULocationID= yt2.DOLocationID 
                    where yt1.RateCodeId=4 """).show()
```

```python
spark.conf.set("spark.sql.adaptive.enabled","true")
spark.conf.set("spark.sql.adaptive.autoBroadcastJoinThreshold","10485760b")
```
